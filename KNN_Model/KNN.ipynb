A- The k-nearest neighbors (KNN) algorithm is a simple, non-parametric supervised machine learning algorithm used for both classification and regression tasks. It operates on the core assumption that similar data points exist in close proximity to one another. 


B- How it Works:
The KNN algorithm does not build a model during a specific training phase (making it a "lazy learner"); instead, it memorizes the entire dataset and performs computations only when a prediction is requested. 
The process for classifying a new data point is as follows:


D-Select K: Choose a positive integer value for k (the number of neighbors to consider).
Calculate Distance: The algorithm calculates the distance between the new data point and all other existing points in the dataset. Common distance metrics include Euclidean distance and Manhattan distance.

C- Find Neighbors: It identifies the k data points with the smallest distances to the new point.
Assign Class/Value:
1- For Classification: The new point is assigned the class label that is most frequent among its k nearest neighbors (a majority vote).
2- For Regression: The new point is assigned the average (or weighted average) of the values of its k nearest neighbors.

```python
def euclidran_distance(x1,x2):
    return np.sqrt(np.sum((x1 -x2)**2))
```

model KNN from scratch

```python
class KNN_model:
    def __init__(self, k = 3): # 3 =>> defult number if him not enter simple
        self.k = k
    def fit(self, x,y):
        self.x_train = x
        self.y_train = y

    def predict(self , X):
        predicted_labels = [self._predict(x) for x in X]
        return np.array(predicted_labels)
    def _predict(x):
        # all distances 
        distances = [euclidran_distance(x, x_train) for x_train in self.x_train]

        # get K nearest sample, labels
        k_indices = np.argsort(distances)[:self.k]
        k_labels = [self.y_train[i] for i in k_indices]

        #get common class label 
        most_common = Counter(k_labels).most_common(1)
        return most_common[0][0]
```

```python
data_iris = datasets.load_iris()
X,y = data_iris.data, data_iris.target
x_train , x_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1234)

print(x_train.shape)
print(y_train.shape)
```

```python
classfier_model = KNN_model(k=3)
classfier_model.fit(x_train, y_train)
```

```python
accurcy = np.sum(y_test) / len(y_test) * 100
print(accurcy)
```
